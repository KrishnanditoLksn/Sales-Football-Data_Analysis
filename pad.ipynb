```python
import pandas as pd
import os
import  glob
from matplotlib import pyplot as plt
path = r'D:\DATASET SALES'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_frame = pd.concat( dfs,axis=0, ignore_index=True)
big_frame.to_csv('dataPenjualanSetahunn.csv')
big_frame.dropna(subset=['Quantity Ordered' , 'Price Each' , 'Order Date' ,'Purchase Address'])
big_frame
```

```python
big_frame['Quantity Ordered'] =pd.to_numeric(big_frame['Quantity Ordered'] , errors='coerce')
big_frame.dropna(inplace=True)
big_frame['Quantity Ordered'] = big_frame['Quantity Ordered'].astype('int32' , errors='ignore')
big_frame['Price Each'] = big_frame['Price Each'].astype('float64' , errors='ignore')
big_frame['Order ID'] = big_frame['Order ID'].astype('int64' , errors='ignore')
```

```python
big_frame['Month'] = big_frame['Order Date'].str[0:2]
big_frame['Month'] = big_frame['Month'].astype('int32')
big_frame.dtypes
```

```python
def get_city(address):
            return address.split(",")[1].strip(" ")
def get_state(address):
            return address.split(",")[2].split(" ")[1]
big_frame['City'] = big_frame['Purchase Address'].apply(lambda x: f"{get_city(x)}  ({get_state(x)})")
big_frame['City'] = big_frame['City'].astype(str , errors='ignore')
big_frame.groupby('City')['City'].count()
big_frame
```

```python
big_frame['Sales'] = big_frame['Quantity Ordered'] *big_frame['Price Each']
big_frame
```

```python
from matplotlib import  pyplot as plt
months = range(1,13)
city = range(1,11)
plt.bar(months , big_frame.groupby('Month')['Sales'].sum())
plt.xticks(months)
plt.ylabel('SALES IN USD')
plt.xlabel('Month')
plt.show()

plt.bar(big_frame.groupby('City')['City'].count().index,big_frame.groupby('City')['City'].count())
plt.xticks(rotation='vertical')
plt.ylabel('SALES IN USD')
plt.xlabel('City')
plt.show()
```

```python
big_frame.groupby('Product')['Quantity Ordered'].sum().plot(kind = 'bar' , color='g')
```

```python
#Overlaying chart masih error
sum_prod = big_frame.groupby('Product').sum()['Quantity Ordered']
quant = big_frame.groupby('Product').sum()['Quantity Ordered']
prices =  big_frame.groupby('Product').mean()['Price Each']
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(sum_prod, quant, color='g')
ax2.plot(sum_prod, prices, color='b')
ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(sum_prod, rotation='vertical', size=8)
fig.show()
```
